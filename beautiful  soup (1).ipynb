```python
cn=soup.select('.link_display_like_text') #company name
location=soup.select('.location_link') #location
# detail=soup.select('.view_detail_button')
table=soup.select('.table') 
```

```python
vd_lst=[]
for link in soup.find_all('a',attrs={'class':'view_detail_button'} ,href=True): #finds links for corresponding 
    linc="https://internshala.com"+link['href']
    vd_lst.append(linc) #calling function and sending linc of each veiw details one by one
    print(linc)
    
```

```python
def VD_get_site(linc): #called function and trying to fetch site name if available else pass
    try:
        req=requests.get(linc)
        soup_VD=bs4.BeautifulSoup(req.text,'lxml')
        #soup_VD
        temp=soup_VD.find_all('a',attrs={'rel':'nofollow noopener noreferrer'} ,href=True)
        temp_str=temp[0].getText()
        print(temp_str)
        return(temp_str)
    except:
        pass
```

```python
for i in range(100):
    sub_list=[]
    try:
        sub_list.append(cn[i].getText())
        sub_list.append(location[i].getText())
        sub_list.append(table[i].getText())
#         sub_list.append(ppo[i].getText())
        website=VD_get_site(vd_lst[i]) ##calling function and sending linc of each veiw details one by one
        sub_list.append(website)
        main_list.append(sub_list)
    except:
        pass
```

```python
import pandas as pd
df = pd.DataFrame(main_list) 
df
```

```python

rows=df.shape[0]
columns=df.shape[1]
```

```python
for i in range (rows):
    for j in range(columns):
        try:
            df.iloc[i,j]=df.iloc[i,j].strip(' ')
            df.iloc[i,j]=df.iloc[i,j].strip('\n')
            df.iloc[i,j]=df.iloc[i,j].replace(" ", "")
            df.iloc[i,j]=df.iloc[i,j].replace(", ", "")
            
            df.iloc[i,j]=df.iloc[i,j].replace("\n", " ")
            df.iloc[i,j]=df.iloc[i,j].replace("StartDate Duration Stipend PostedOn ApplyBy      ","")

    #         df.iloc[i,j]=" ".join(df.iloc[i,j].strip(' '))
        except:
            pass


df.columns = ['Name','Location','table','website']
# df
```

```python
# str=df.iloc[1,2]
# str.replace("\n", " ")
# str.replace("StartDate Duration Stipend PostedOn ApplyBy      ","")


```

```python
tables=list(df.table)
applyBy=[]
Duration=[]
Stipend=[]
for i in range(len(tables)):
    detail=tables[i].split(" ")
    for ele in detail: 
        if ele == '': 
            detail.remove(ele)
    applyBy.append(detail[0])
    Duration.append(detail[1])
    Stipend.append(detail[2])
df.insert(3,"Duration",Duration)
df.insert(4,"Stipend",Stipend)
df=df.drop(['table'], axis = 1)
```
